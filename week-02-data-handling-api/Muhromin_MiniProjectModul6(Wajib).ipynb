```python
import requests

kota = input("Masukkan nama kota/kabupaten (contoh: KOTA BANDUNG): ").upper()
url = "https://api-sekolah-indonesia.vercel.app/sekolah/smk?perPage=1000"

res = requests.get(url).json()
data = res["dataSekolah"]

hasil = [s["sekolah"] for s in data if kota in s["kabupaten_kota"].upper()]

print(f"\nDaftar SMK di {kota}:")
for i, nama in enumerate(hasil, 1):
    print(f"{i}. {nama}")

print(f"\nTotal: {len(hasil)} sekolah ditemukan.")
```
